# Is a *square wave* a sum of trig functions?


The *square wave* on $[-1/2, 1/2]$ it looks like this

$$
\mathrm{Sq}(x) = \begin{cases}-1 &\text{if } x<0 \\ 1 &\text{if } x>0 \end{cases}
$$

**Goal:** Find numbers $a_n$ such that 
$$
b_1 \sin(2π\, x)+b_2 \sin(2π\, 2x)+\ldots + b_N \sin(2π\, N x)
$$
is "close" to $\mathrm{Sq}(x)$.


```python

@NamedFunction
def Sq(x):
    return sign(sin(2*pi*x))

fig_square = Figure(x_range=(-1 / 2, 1 / 2), y_range=(-2, 2))
fig_square.title = f"The square wave $\\mathrm{{Sq}}(x)$"
fig_square.show()

Nmax=10
model=sum([b[n]*sin(2*pi*n*x) for n in range(1,Nmax+1) ])

with fig_square:
    for n in range(1,Nmax+1):
        parameter(b[n],min=-1.5,max=1.5,value=0)
    plot(Sq(x),x, id="Sq", label=r'$\operatorname{Sq}(x)$')
    plot(model, x, id="trig_sum", label=r'trig sum')
  
    
```
