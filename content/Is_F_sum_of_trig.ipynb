# A function as a sum of sines

```python

from gu_toolkit import *
from helpers.Fourier_01_helper import *
```

```python

# Nmax - the number of *modes* 
Nmax = 4
F = create_mystery_function(Nmax)
```

```python

fig = Figure(x_range=(-1 / 2, 1 / 2), y_range=(-2.5, 2.5))
display(
    Markdown(
        f"""Here is a random function $F(x)$. Can you represent it as a sum of trig functions? 
        $$
        F(x)= b_1 \\sin(2\\pi \\cdot x)+ b_2 \\sin(2\\pi\\cdot 2 x)+\\ldots +b_{{ {Nmax} }} \\sin(2\\pi\\cdot {Nmax} x)
        $$
        \n\n\n with random numbers $b_n$ between $-1$ and $1$."""
    )
)

fig.show()
model=sum([b[n]*sin(2*pi*n*x) for n in range(1,Nmax+1) ])
with fig:
    set_title( r"Is $F(x)$ a sum of trig functions?")
    plot(F(x), x, id="F(x)", label=r'$F(x)$')
    plot(model, x, id="trig_sum", label=r'trig sum')
```
